```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_largest(a, n):
    # n must be strictly positive
    assert n > 0
    a_sort = np.sort(a)
    return a_sort[-n:]
```

```python
a = np.array([1,4,2,4.3,5, 12, 7, 8,2])
get_largest(a, 5)
```

```python
def unique_rows(A):
    return np.unique(A, axis=0)
```

```python
def cauchy_matrix(x, y):
    #transpose x to a column vector
    x = x.reshape((-1,1))
    return 1/(x - y)
```

```python
x = np.array([1,2,3])
y = np.array([4,5,6])
cauchy_matrix(x,y)
```

```python
def rose_curve(n):
    theta = np.linspace(0, 2*np.pi, num=1000)
    rho = np.cos(n*theta)
    x = rho*np.cos(theta)
    y = rho*np.sin(theta)
    return x,y

def plot_roses(n):
    x,y = rose_curve(n)
    plt.plot(x,y)
    plt.show()
```

```python
plot_roses(4)
```

```python
plot_roses(6)
```

```python
plot_roses(7)
```

```python
def first_occurrence(x, y):
    occurrences = np.where(x==y)[0]
    return occurrences[0] if len(occurrences) > 0 else -1
```

```python
x = np.array([1,2,4,21,5,2,3])
y = 2
first_occurrence(x, y)
```
