```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.uniform(low=-5, high=5, size=(1000,2))
padding = 0.3
x[x>0] += padding
x[x<0] += -padding
noise = 0.3*np.random.uniform(low=-5, high=5, size=(1000,))
y = (np.sign(x[:,0]+ noise)==np.sign(x[:,1]+noise)).astype(float)
plt.scatter(x[:,0],x[:,1],c=y)
plt.xlabel('X1')
plt.ylabel('X2')
plt.show()

dataset = np.column_stack((x, y))  # Combine features and labels
np.savetxt('xor.txt', dataset, delimiter=',')
```
